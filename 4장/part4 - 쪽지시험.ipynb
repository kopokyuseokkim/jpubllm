**질문 1.**  spaCy를 활용하여 다음 문장을 품사 태깅을 해보세요.
 “The way to get started is to quit talking and begin doing.”

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("The way to get started is to quit talking and begin doing.")

for token in doc:
    print(token.text, token.pos_)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("The way to get started is to quit talking and begin doing.")

for token in doc:
    print(token.text, token.lemma_)
```

**질문 4.** Okt 형태소 분석기로 위 문장을 다음의 불용어 리스트를 활용하여 불용어를 제거하세요.
 stopwords = ["을", "를", "이", "가", "은", "는", "에", "한", "하다"]

```python
from konlpy.tag import Okt

okt = Okt()
text = "별을 노래하는 마음으로 모든 죽어가는 것을 사랑해야지."
tokens = okt.morphs(text)

stopwords = ["을", "를", "이", "가", "은", "는", "에", "한", "하다"]

filtered_tokens = [word for word in tokens if word not in stopwords]
print(filtered_tokens)
```

**질문 5.**  모듈Okt 모듈을 활용하여 문장을 입력하면 명사만 리스트로 반환하는 모듈을 만드세요.
 ex) 나는 소년이다
['나', '소년']

```python
from konlpy.tag import Okt

def getNouns(line):
    okt = Okt()
    results = okt.nouns(line)
    return results

getNouns('나는 소년이다')
```
